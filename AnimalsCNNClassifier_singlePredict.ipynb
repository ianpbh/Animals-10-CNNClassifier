```python
import tensorflow as tf
import numpy as np
from PIL import Image
```

```python
#Change the string below to be your model checkpoint
#Altere a string abaixo para ser o checkpoint do seu modelo
model = tf.keras.models.load_model('trained_model/AnimalsClassifierCheckpoint40.0.89.h5')
```

```python
#Representação numérica de cada classe, dict copiado de "train_data.class_indices", e traduzido para inglês
#Numerico representation of each class, dict copied from "train_data.class_indices", and translated to english
classes = {'dog': 0,
           'horse': 1,
           'elephant': 2,
           'butterfly': 3,
           'chicken': 4,
           'cat': 5,
           'cow': 6,
           'sheep': 7,
           'spider': 8,
           'squirrel': 9}
```

```python
#The code below import the image
#O código abaixo importa a imagem
from PIL import Image
image = Image.open('spider.jpeg')

#Resize it to the shape our CNN expects
#Redimensiona para o formato da CNN
image = image.resize((224,224))

#Convert it to numpy array, and expand it's dimension to add the batch size
#Converte para numpy array, e expande a dimensão para adicionar o tamanho do batch
numpydata = np.asarray(image)
numpydata = numpydata / 255.0
numpydata = np.expand_dims(numpydata, axis=0)

#make the prediction
#Faz a predição
prediction = model.predict(numpydata)

#Cria um dict para visualizar o nome da classe predita
#Create a dict to visualize the predicted class name
index_to_class = {v: k for k, v in classes.items()}
print(index_to_class[np.argmax(prediction)])
image
```
